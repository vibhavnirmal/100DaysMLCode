```
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import random
import tensorflow as tf
from google.colab.patches import cv2_imshow
```

```
legs = ['steel', 'wood']
back = ['Febric', 'Wood']
seat = ['Fabric', 'Leather', 'Wood']
couch = ['Fabric', 'Leather','Valvet', 'Wood']
IMG_SIZE = 128
```

```
leg_model = tf.keras.models.load_model('drive/My Drive/Lag_model.model')
seat_model = tf.keras.models.load_model('drive/My Drive/Seat_model.model')
back_model = tf.keras.models.load_model('drive/My Drive/Back_model.model')
couch_model = tf.keras.models.load_model('drive/My Drive/Final_couch_model.model')
```

```
from google.colab import files
uploaded = files.upload()
```

```
for name in uploaded.keys():
  image = cv2.imread(name)
  image = cv2.resize(image, (IMG_SIZE, IMG_SIZE))
  cv2_imshow(image)
  image = image.reshape(-1, IMG_SIZE, IMG_SIZE, 3)
  couch_prediction = couch_model.predict(image)
  back_prediction = back_model.predict(image)
  seat_prediction = seat_model.predict(image)
  leg_prediction = leg_model.predict(image)

  print("Couch - ", couch[couch_prediction[0].argmax()])
  print("Back - ", back[back_prediction[0].argmax()])
  print("Seat - ", seat[seat_prediction[0].argmax()])
  print("Legs - ", legs[leg_prediction[0].argmax()])
  print("----------------------------------------------------------------------")
```
